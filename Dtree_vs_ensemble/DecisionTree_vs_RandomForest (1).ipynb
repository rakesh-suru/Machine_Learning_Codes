```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("Mobileprice.csv")
```

```python
X = data.drop('price_range', axis=1)
Y = data['price_range']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=16)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100, random_state=16)

rfc.fit(X_train, Y_train)

rfc_pred = rfc.predict(X_test)

rfc_acc = accuracy_score(Y_test, rfc_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(criterion= "entropy")

dtree.fit(X_train, Y_train)

dtree_pred = dtree.predict(X_test)

dtree_acc = accuracy_score(Y_test, dtree_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
dtree_cm = confusion_matrix(Y_test, dtree_pred)
dtree_cm
```

```python
rfc_cm = confusion_matrix(Y_test, rfc_pred)
rfc_cm
```

```python
from sklearn.metrics import classification_report
```

```python
dtree_report = classification_report(Y_test, dtree_pred)
print(dtree_report)
```

```python
rfc_report = classification_report(Y_test, rfc_pred)
print(rfc_report)
```

```python
print(f"Decision Tree Accuracy: {dtree_acc}")
print(f"Random Forest Accuracy: {rfc_acc}")


sns.heatmap(dtree_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Decision Tree Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

sns.heatmap(rfc_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
models = ['Decision Tree', 'Random Forest']
accuracies = [dtree_acc, rfc_acc]

plt.figure(figsize=(6, 5))
plt.bar(models, accuracies, color=['red', 'blue'])
plt.ylabel('Accuracy')
plt.title('Comparison of Decision Tree and Random Forest Accuracies')
plt.grid(axis= "y")
plt.show()
```
