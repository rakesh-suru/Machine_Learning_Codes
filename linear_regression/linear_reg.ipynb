```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
USA["Address"] = le.fit_transform(USA["Address"])
```

```python
USA.info()
```

```python
USA.head()
```

```python
USA.columns
```

```python
X = USA[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population', 'Address']]
Y = USA["Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 44)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
prediction = lm.predict(X_test)
```

```python
plt.scatter(range(len(Y_test)), Y_test, color = "red", alpha = 0.6, label = "Actual Values")
plt.scatter(range(len(prediction)), prediction, color = "green", alpha = 0.6, label = "Predicted Values")
plt.legend()
```

```python
from sklearn import metrics
```

```python
print("MAE :", metrics.mean_absolute_error(Y_test, prediction))
```

```python
print("MSE :", metrics.mean_squared_error(Y_test, prediction))
```

```python
print("RMSE :", metrics.root_mean_squared_error(Y_test, prediction))
```

```python
print("R2_score :", metrics.r2_score(Y_test, prediction))
```
